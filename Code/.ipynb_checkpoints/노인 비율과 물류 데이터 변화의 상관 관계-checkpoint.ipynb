```python
#필요한 모듈 추가
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
#데이터 불러오기
df_log=pd.read_excel('2020 빅콘테스트_CJ올리브네트웍스_제공DB.xlsx')
df_log
```

```python
#서울데이터만 추출 
df_seoul=df_log[df_log['CTPV_NM']=='서울특별시'].loc[:,['DL_YMD','HDNG_NM','INVC_CONT']]
df_seoul
```

```python
#칼럼 이름 바꾸기
df_seoul=df_seoul.rename(columns={"DL_YMD":"날짜","HDNG_NM":"동","INCV_CONT":"송장건수"})
df_seoul
```

```python
#2019년,2020년 2월-5월 동별 물류 건수 
df=pd.pivot_table(data=df_seoul, index=['날짜'],columns=['동'], aggfunc=['sum'])
df=df.fillna(0)#결측치는 0건으로 표시
df
```

```python
#2019,2020년도 동별 총 물류 건수 구하기

df_2019=df.iloc[:120,:] #2019년 동별 총 물류 건수
df_2019=df_2019.sum(axis=0)

df_2020=df.iloc[120:241] #2020년 동별 총 물류 건수
df_2020=df_2020.drop(200229,axis=0) #2019년과 2020년 비교를 위해 2020년 2월 29일 데이터 삭제 
df_2020=df_2020.sum(axis=0)

df_sum=pd.concat([df_2019,df_2020],axis=1) #2019년 합계와 2020년 합계 데이터 합치기  
df_sum.columns=['2019','2020']
df_sum
```

```python
#연간 동별 물류 건수 변화율 
df_sum["변화율(%)"]=(df_sum['2020']/df_sum['2019']-1)*100
df_roc=df_sum[['변화율(%)']]
df_roc=df_roc.rename(index={'HDNG_NM':'동'})
df_roc
```

```python
#서울시 동별 노인 인구 비율
df_elder=pd.read_excel('서울시_동별인구_노원구,중구.xlsx')
df_elder["65세이상인구비율"]=df_elder['65세이상인구']/df_elder['전체인구']*100
df_elder=df_elder.sort_values(by='동',
                     axis=0,
                     ascending=True)#동 이름 가나다 순으로 배열
df_elder=df_elder[['동','65세이상인구비율']]#column 이름 설정
df_elder=df_elder.set_index("동")
df_elder=df_elder.rename(index={"신당5동":"신당제5동"})#df_roc와 index 값 통일 
df_elder
```

```python
#연간 동별 물류 건수 변화율과 노인 인구 비율 데이터 합치기 
df_result = pd.merge(df_roc,df_elder,how='outer', on='동')
df_result.columns=['물류 송장 건수 변화율(%)','65세 이상 인구 비율(%)']
df_result
```

```python
# 두 변수의 상관관계 시각화 하기 
import matplotlib as mpl
%matplotlib inline

from matplotlib import font_manager,rc #한글 폰트 사용
font_location = 'C://Windows//Fonts//malgun.ttf'
font_name=font_manager.FontProperties(fname=font_location).get_name()
mpl.rc('font',family=font_name)

sns.pairplot(df_result, kind='reg',height=4)
```

```python
#두 변수의 상관계수 구하기 
cor_result=stats.pearsonr(df_result['물류 송장 건수 변화율(%)'],df_result['65세 이상 인구 비율(%)'])
cor_result
#피어슨 상관계수=-0.38 -> 뚜렷한 음적 상관관계가 있음
#'65세 이상 인구 비율'과 '물류 송장 건수 변화율' 사이에는 음의 상관 관계가 있음 
```
